```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('/kaggle/input/housing/housing.csv')
```

```python
data
```

```python
data.info()
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split 

x = data.drop(['median_house_value'] , axis =1)
y = data['median_house_value']
```

```python
x
```

```python
y
```

```python
#splitting our data 

x_train ,x_test ,y_train  , y_test = train_test_split(x , y ,  test_size = 0.2)
```

```python
y_train
```

```python
train_data = pd.concat([x_train ,y_train] , axis=1)
```

```python
train_data
```

```python
train_data.hist(bins=20, figsize=(10, 8))  # Increase the figure size (adjust as needed)
```

```python
# Heatmap for data correlation

# Select only numeric columns for correlation matrix
numeric_cols = train_data.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 10))
sns.heatmap(numeric_cols.corr(), annot=True, cmap="Blues")
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms']+1)
train_data['total_rooms'] =np.log(train_data['total_rooms']+1)
train_data['population'] = np.log(train_data['population']+1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms']+1)

```

```python
train_data.hist(bins=20, figsize=(12, 12))  # Increase the figure size (adjust as needed)
```
